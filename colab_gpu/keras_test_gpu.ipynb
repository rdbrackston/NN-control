# Run with small number of hidden units

```
N, D_in, H, D_out = 64, 1024, 128, 8

X = np.random.normal(0,1,(N, D_in))
y = np.random.normal(0,1,(N, D_out))

#  define the keras model
model = Sequential()
model.add(Dense(H, input_dim=D_in, activation='relu'))
model.add(Dense(D_out, activation='relu'))
model.add(Dense(D_out, activation='sigmoid'))
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
start_time = time.time()
model.fit(X, y, epochs=3, batch_size=128, verbose=0)
print("runtime: ", time.time()-start_time)
```

# Use large number of hidden units

```
N, D_in, H, D_out = 64, 10000, 10000, 8

X = np.random.normal(0,1,(N, D_in))
y = np.random.normal(0,1,(N, D_out))

#  define the keras model
model = Sequential()
model.add(Dense(H, input_dim=D_in, activation='relu'))
model.add(Dense(D_out, activation='relu'))
model.add(Dense(D_out, activation='sigmoid'))
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
start_time = time.time()
model.fit(X, y, epochs=10, batch_size=256, verbose=0)
print("runtime: ", time.time()-start_time)
```
